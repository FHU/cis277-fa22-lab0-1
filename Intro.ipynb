## While Loop (less Pythonic)

```python
s = "RacecAr "

def is_palindrome1(string):
    clean_string = string.lower().replace(' ', '')
    first, last = 0, len(clean_string)-1
    while( first < last):
        if clean_string[first] == clean_string[last]:
            first +=1
            last -=1
        else:
            return False
    
    return True

is_palindrome1(s) 
```

## String Slicing (More Pythonic)

```python
def is_palindrome2(string):
    clean_string = string.lower().replace(' ', '')
    reversed_string = clean_string[::-1]
    return clean_string == reversed_string

s = " Rac eC ar "
is_palindrome2(s)
```

```python
import csv

ratings = {}

with open('movie.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        #print(row['movie_title'], row['imdb_score'])
        ratings[row['movie_title']] = row['imdb_score']
```

```python
values_list = []

with open('movie.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        values = float(row['imdb_score'])
        values_list.append(values)

average_value = sum(values_list) / len(values_list)
        


```

```python
max_value = max(values_list)

with open('movie.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if row['imdb_score'] == str(max_value):
            best_movie = row['movie_title']
```

```python
min_value= min(values_list)

with open('movie.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if row['imdb_score'] == str(min_value):
            worst_movie = row['movie_title']
```

```python
star_wars_movies = []

with open('movie.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if 'Star Wars' in row['movie_title']:
            star_wars_movies.append(row['movie_title'])
```

```python
director_dictionary = {}

with open('movie.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if row["director_name"] in director_dictionary:
            director_dictionary[row["director_name"]] += 1
        else:
            director_dictionary[row["director_name"]] = 1
```

```python
duration_list = []

with open('movie.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if "Action" in row["genres"]:
            if row["duration"] !=  '':
                duration_values = float(row["duration"])
                duration_list.append(duration_values)

average_runtime = sum(duration_list) / len(duration_list)
        



```
